```python
filt = df['Country'] == 'India'
df.loc[filt]['LanguageHaveWorkedWith'].str.contains('Python').sum()
```

```python
country_users_python = countrygrp['LanguageHaveWorkedWith'].apply(lambda x: x.str.contains('Python').sum())
country_users_python
```

```python
country_respondents = df['Country'].value_counts()
country_respondents
```

```python
python_df = pd.concat([country_respondents,country_users_python], axis = 'columns', sort = False)
python_df.rename( columns = {'count' : 'NumRespon', 'LanguageHaveWorkedWith' : 'PythonUsersPerCountry'}, inplace = True)
python_df
```

```python
python_df['PctKnowsPython'] = (python_df['PythonUsersPerCountry']/python_df['NumRespon'])*100
python_df
```

```python
filt_1 = python_df['NumRespon'] > 50
python_df[filt_1].sort_values(by = 'PctKnowsPython', ascending = False)
```
